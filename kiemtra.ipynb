```
#1
listA = [1, 10, 20, 5, 60, 30, 10, 9, 10]
listB = []
for x in listA:
  if x!=10:
    listB.append(x)
print("kết quả câu 1: ", listB)
```

```
#2
list1 = ["A", "B", "C", "D", "E", "F"]
list2 = [1,2,3,4,5,6]
Dict = dict(zip(list1, list2))
print("kết quả câu 2: ",Dict)
```

```
#3
person={
			"Name":"Quoc Nam",
			"Age": 28,
			"Salary":8000,
			"City":"Tuy Hoa"
	}
KeyRemove=['Salary','City']
for x in KeyRemove:
		if x in person:
			person.pop(x)
print("kết quả câu 3:",person)
```
